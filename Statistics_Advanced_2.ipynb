Question 1: What is hypothesis testing in statistics?

Answer: Hypothesis Testing in Statistics

Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data.

It helps us check whether a claim (or assumption) about a population parameter is likely to be true.

Steps in Hypothesis Testing

State the hypotheses

Null Hypothesis (H₀): Assumes no effect or no difference.
Example: "The average height of students = 170 cm."

Alternative Hypothesis (H₁ or Ha): Assumes there is an effect or difference.
Example: "The average height of students ≠ 170 cm."

Choose a significance level (α)

Common choices are 0.05 or 0.01.

This is the probability of rejecting H₀ when it is actually true (Type I error).

Select a test statistic

Example: z-test, t-test, chi-square test, ANOVA depending on data type.

Calculate the p-value or critical value

p-value: Probability of obtaining results at least as extreme as the sample, assuming H₀ is true.

Make a decision

If p-value ≤ α, reject H₀ (evidence supports H₁).

If p-value > α, fail to reject H₀ (not enough evidence against H₀).

Example

Suppose a company claims that the average life of its batteries is 100 hours.

H₀: μ = 100

H₁: μ ≠ 100

You test a sample of 30 batteries. If your statistical test gives a p-value = 0.03 and α = 0.05 → Reject H₀.

Conclusion: The average battery life is significantly different from 100 hours.

Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

Answer: Significance Level (α) in Hypothesis Testing

The significance level (α) is the threshold probability used to decide whether to reject the null hypothesis (H₀).

It represents the risk of making a Type I error → rejecting H₀ when it is actually true.

Common choices:

α = 0.05 (5%) → Most widely used.

α = 0.01 (1%) → Stricter (used in medical/critical studies).

α = 0.10 (10%) → More lenient (used in exploratory research).



Role in Hypothesis Testing

Set before testing → Decide α before collecting data to avoid bias.

Compare with p-value:

If p-value ≤ α → Reject H₀ (significant evidence for H₁).

If p-value > α → Fail to reject H₀ (insufficient evidence).

Controls confidence:

Significance level is linked to confidence level.

Confidence level = 1 − α.

Example: α = 0.05 → 95% confidence level.


Example

A company claims its new bulb lasts 1000 hours.

H₀: μ = 1000

H₁: μ ≠ 1000

Significance level: α = 0.05

Suppose test results → p-value = 0.03.

Since 0.03 < 0.05, we reject H₀.

Conclusion: The bulb life is significantly different from 1000 hours.


Key Idea

The significance level (α) sets the "cut-off line" for making decisions:

Lower α → more strict, harder to reject H₀.

Higher α → less strict, easier to reject H₀ (but higher risk of false positives).




Question 4: What are Type I and Type II errors? Give examples of each.

Answer: Type I and Type II Errors

When we do hypothesis testing, there are 2 possible decisions:

Reject H₀

Fail to reject H₀

Since we are working with sample data (not the full population), errors can happen.

1. Type I Error (False Positive)

Happens when we reject H₀ even though it is true.

Probability of making this error = α (significance level).

Example:

Null Hypothesis (H₀): A man is not guilty of a crime.

Alternative (H₁): The man is guilty.

Type I Error: The jury says he is guilty (rejects H₀) when he is actually innocent.

Real-world example:

A new medicine is tested. H₀: "The medicine has no effect."

If we reject H₀ but in reality the medicine has no effect, we commit a Type I error (false claim).


2. Type II Error (False Negative)

Happens when we fail to reject H₀ even though it is false.

Probability of making this error = β.

Power of a test = 1 − β (probability of correctly rejecting H₀).

Example:

Null Hypothesis (H₀): A man is not guilty.

Alternative (H₁): The man is guilty.

Type II Error: The jury says he is not guilty (fails to reject H₀) when he is actually guilty.


Real-world example:

H₀: "The new medicine has no effect."

If we fail to reject H₀ but in reality the medicine works, we commit a Type II error (missed discovery).



Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.

Answer: Z-test vs T-test in Statistics

Both are hypothesis tests used to compare sample statistics with population parameters, but they differ based on sample size and variance knowledge.

1. Z-test

Used when:
Sample size is large (n > 30)
Population variance (σ²) or standard deviation (σ) is known
Data is normally distributed

Test statistic formula:

Z = x-μ/σ/σ/n

Example:
A company claims the average salary = ₹50,000.

σ = ₹5,000 (population standard deviation known)

n = 100 employees surveyed
Use Z-test.


2. T-test

Used when:
✅ Sample size is small (n ≤ 30)
✅ Population variance (σ²) is unknown (we use sample variance s² instead)
✅ Data is approximately normally distributed

Test statistic formula:

T = X - μ / s/n
	​


Example:
A teacher wants to test if the average marks of 25 students = 60.

σ unknown, only sample standard deviation available.
 Use T-test.
